# Questão 2

Leia o arquivo `population.csv.` Coloque esta tabela no formato `Tidy`. Leia o arquivo `tuberculosis.csv`.
Crie um ranking dos **cinco piores pares (país, ano)**, **ordenado pelo número de casos de tuberculose por 10.000 habitantes**. Descarte linhas com valores nulos e exiba os resultados com uma casa decimal.

````python
#Importar arquivo population
pop =  pd.read_csv('/content/population.csv', thousands='.')

#Transformar Wide Dataframe Population para Tidy DF
n_pop = pop.melt('country', var_name = 'year', value_name = 'population')

#Transformar coluna Year para Int64 para realizar o Join 
n_pop['year'] = pd.to_numeric(n_pop['year'])

n_pop.head()
````

````python
tub = pd.read_csv('/content/tuberculosis.csv')
tub.head()
````

````python
#Fazer um merge dos datasets Population & Tuberculosis
novo_dataset = pd.merge(n_pop, tub, how='left', on=['country', 'year'])
#Substituir os NaN's por 0
novo_dataset.fillna(0, inplace=True)
#Verificar Populacao Tuberculosa que veio da base 'Tuberculosis.csv'
novo_dataset['populacao_TUB'] = novo_dataset['child'] + novo_dataset['adult'] + novo_dataset['elderly']

#Número de Casos / Número da Populacao * 10.000
novo_dataset['Casos_Tuberculose_Prevalencia_Epidemiológica'] = round(((novo_dataset['populacao_TUB'] / novo_dataset['population'])*10000),1)
#Mostrar os piores países e anos com maiores casos de tuberculose
novo_dataset.sort_values(by='Casos_Tuberculose_Prevalencia_Epidemiológica', ascending = False).head()


````

#Questão 3

Leia os arquivos `flights.csv` e `airlines.csv`. Para cada companhia aérea, **defina dois níveis de atraso médio de partida** `dep_delay`: **menor ou igual a 12.64 minutos (Low) e acima de 12.64 minutos (High)**. Exiba o nome da companhia, o número de vôos e o nível do atraso médio de partida.

````python
flights = pd.read_csv('/content/flights.csv')
flights.head()
````

````python
airlines = pd.read_csv('/content/airlines.csv')
airlines.head()
````

````python
#Concatenar Dataframes Flights & Airlines
flights_airlines_concat = pd.concat([flights, airlines], ignore_index = True)

#Criar uma função para dar nome a companhia baseado no código de Carrier
def nomeCompanhia(row):
  if row['carrier'] == '9E':
    return 'Endeavor Air Inc.'
  elif row['carrier'] == 'AA':
    return 'American Airlines Inc.'
  elif row['carrier'] == 'AS':
    return 'Alaska Airlines Inc.'
  elif row['carrier'] == 'B6':
    return 'JetBlue Airways'
  elif row['carrier'] == 'DL':
    return 'Delta Air Lines Inc.'
  elif row['carrier'] == 'EV':
    return 'ExpressJet Airlines Inc.'
  elif row['carrier'] == 'F9':
    return 'Frontier Airlines Inc.'
  elif row['carrier'] == 'FL':
    return 'AirTran Airways Corporation'
  elif row['carrier'] == 'HA':
    return 'Hawaiian Airlines Inc.'
  elif row['carrier'] == 'MQ':
    return 'Envoy Air'
  elif row['carrier'] == 'OO':
    return 'SkyWest Airlines Inc.'
  elif row['carrier'] == 'UA':
    return 'United Air Lines Inc.'
  elif row['carrier'] == 'US':
    return 'US Airways Inc.'
  elif row['carrier'] == 'VX':
    return 'Virgin America'
  elif row['carrier'] == 'WN':
    return 'Southwest Airlines Co.'
  elif row['carrier'] == 'YV':
    return 'Mesa Airlines Inc.'

def atrasoMedio_Status(row):
#A média do dep_delay do DF é 12.64 - 
  if row['dep_delay'] <= 12.64:
     return 'Low'
  else:
     return 'High'

flights_airlines_concat['name'] = flights_airlines_concat.apply(nomeCompanhia, axis = 1)
flights_airlines_concat['status'] = flights_airlines_concat.apply(atrasoMedio_Status, axis = 1)

flights_airlines_concat.head()
````

````python
#Verificar se há valores ausentes no DF
checkar_valores_nan = flights_airlines_concat.isnull().any().any()
print(checkar_valores_nan)
````

````python
flt_airlines_copia = flights_airlines_concat.dropna()
flt_airlines_copia.head()
````

````python
#Número de Voos realizados pelas companhias - também pode ser feito pela coluna carrier
carrier_cnts = flt_airlines_copia['name'].value_counts()
carrier_cnts
````

### a.
Faça a união das duas tabelas de quatro maneiras, correspondentes aos "joins": left, right, inner e outer. Imprima os resultados obtidos de cada uma das maneiras.

````python
gyms = pd.DataFrame({"gym": ["BodyTech", "Smartfit", "Bluefit", "Unio Hot Yoga", "Villa Pilates"],
                       "units": [3, 45, 20, 2, 1]})

monthly_used = pd.DataFrame({"clients": ["Pedro", "Jenny", "Pedro", "Joaquim", "Joaquim"],
                        "gyms_used": ["BodyTech", "Smartfit", "Bluefit", "Unio Hot Yoga", "Smartfit"],
                        "monthly_frequency": [4, 8, 4, 2, 4]})
````

````python
gyms
````

````python
monthly_used
````

````python
gyms.dtypes
````

````python
monthly_used.dtypes
````

````python
inner = pd.merge(gyms, monthly_used, how='inner', left_on=['gym'], right_on=['gyms_used'], indicator = 'qual_coluna')
inner
````

````python
right = pd.merge(gyms, monthly_used, how='right', left_on=['gym'], right_on=['gyms_used'], indicator = 'qual_coluna')
right
````

````python
left = pd.merge(gyms, monthly_used, how='left', left_on=['gym'], right_on=['gyms_used'], indicator = 'qual_coluna')
left
````

````python
outer = pd.merge(gyms, monthly_used, how='outer', left_on=['gym'], right_on=['gyms_used'], indicator = 'qual_coluna')
outer
````

Explique cuidadosamente na célula abaixo as diferenças nos resultados produzido pelos quatro tipos de união. Qual dos quatro "joins" é mais inclusivo? Qual é mais exclusivo?

### Resposta Questão 5 B



> O Outer Join sempre vai ser mais **inclusivo**, *pois ele vai pegar um maior número de dados* (nesse caso das duas tabelas `gym` & `monthly_used`. E teremos também por essa razão teremos que trabalhar um pouco na limpeza desses dados. Por outro lado, o Inner Join será mais **exclusivo** pois ele obterá dados relacionados das duas tabelas que satisfazem a condição de junção de ambas tabelas.

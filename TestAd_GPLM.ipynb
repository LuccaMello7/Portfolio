````python
#Concatenar Dataframes Flights & Airlines
flights_airlines_concat = pd.concat([flights, airlines], ignore_index = True)

#Criar uma função para dar nome a companhia baseado no código de Carrier
def nomeCompanhia(row):
  if row['carrier'] == '9E':
    return 'Endeavor Air Inc.'
  elif row['carrier'] == 'AA':
    return 'American Airlines Inc.'
  elif row['carrier'] == 'AS':
    return 'Alaska Airlines Inc.'
  elif row['carrier'] == 'B6':
    return 'JetBlue Airways'
  elif row['carrier'] == 'DL':
    return 'Delta Air Lines Inc.'
  elif row['carrier'] == 'EV':
    return 'ExpressJet Airlines Inc.'
  elif row['carrier'] == 'F9':
    return 'Frontier Airlines Inc.'
  elif row['carrier'] == 'FL':
    return 'AirTran Airways Corporation'
  elif row['carrier'] == 'HA':
    return 'Hawaiian Airlines Inc.'
  elif row['carrier'] == 'MQ':
    return 'Envoy Air'
  elif row['carrier'] == 'OO':
    return 'SkyWest Airlines Inc.'
  elif row['carrier'] == 'UA':
    return 'United Air Lines Inc.'
  elif row['carrier'] == 'US':
    return 'US Airways Inc.'
  elif row['carrier'] == 'VX':
    return 'Virgin America'
  elif row['carrier'] == 'WN':
    return 'Southwest Airlines Co.'
  elif row['carrier'] == 'YV':
    return 'Mesa Airlines Inc.'

def atrasoMedio_Status(row):
#A média do dep_delay do DF é 12.64 - 
  if row['dep_delay'] <= 12.64:
     return 'Low'
  else:
     return 'High'

flights_airlines_concat['name'] = flights_airlines_concat.apply(nomeCompanhia, axis = 1)
flights_airlines_concat['status'] = flights_airlines_concat.apply(atrasoMedio_Status, axis = 1)

flights_airlines_concat.head()
````

````python
import pandas as pd
from pandasql import sqldf

df = pd.read_csv("/content/tickets.csv", sep=",")
df.head()
````

````python
df.rename(columns={"ticket_created_timestamp":"ticket_time",
                   "requester_external_id":"external_id_rqst",
                   "fcr":"resolved_first_call",
                   "general_tribe": "which_group"}, inplace = True)
df.info()
````

````python
df = df.astype({'resolved_first_call': float, 'ticket_id': int, 'external_id_rqst':int})
df.info()
````

````python

query = """

WITH table questao_6 AS (

SELECT 

ticket_id,
resolved_first_call,
category

from df

where

case when count(strftime('%d-%m-%Y', ticket_time)) BETWEEN '21/02/2022' AND '28/02/2022' THEN 'last_week'
     when count(strftime('%d-%m-%Y', ticket_time)) BETWEEN '13/02/2022' AND '28/02/2022' THEN  'last_15days'
     when count(strftime('%d-%m-%Y', ticket_time)) BETWEEN '07/02/2022' AND '28/02/2022' THEN  'last_21days'
     when count(strftime('%d-%m-%Y', ticket_time)) BETWEEN '28/01/2022' AND '28/02/2022' THEN  'last_30days'
     ELSE 'more_than_30_days'
  
END as 'volume_de_contatos'
)

SELECT 
	 count(ticket_id) as number_of_tickets
 	,SUM(resolved_first_call) as number_of_tickets_resolved_first_call
	,ROUND(100 * sum(resolved_first_call) / count(ticket_id),2) as percentage_resolved_first_call
  ,RANK() OVER (ORDER BY category DESC) AS rank_categorias_com_mais_tickets

FROM questao_6


"""

'''
poderia usar o datediff como no SQLServer - e usar 
o @declare para setar a data inicial de 28/02/2022 
entretanto preferi setar manualmente por questão de tempo
'''

df_copia = sqldf(query)
df_copia
````
